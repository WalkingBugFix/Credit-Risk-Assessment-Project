## A company will not be tagged as a defaulter if its net worth next year is positive, or else, it'll be tagged as a defaulter.

```python
# defining the dependent variable
data['Default'] = data['Networth Next Year'].apply(lambda x: 1 if x<0 else 0)
```

```python
# dropping the redundant features
data.drop(columns=['Num','Networth Next Year','Deposits (accepted by commercial banks)'], axis=1, inplace=True)
```

```python
data.shape
```

```python
#renaming the feature in the data for convenience
data.columns = data.columns.str.replace(' ','_')
```

# Feature Engineering

```python
features = [feature for feature in data.columns if feature != 'Default']
features
```

```python
# replacing the outliers in the data with the nearest non-outlier value
for col in features:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # capping the values to the whiskers or IQR bounds
    data[col] = np.where(data[col] < lower_bound, lower_bound, np.where (data[col] > upper_bound, upper_bound, data[col]))
data.head()
```

```python
# to check how the outlier treatment worked, we will plot the boxplots again
for feature in data.columns:
    sns.boxplot(x = data['Default'], y=feature, data=data, palette='YlGnBu')
    plt.title(f'{feature} vs Dependent Variable')
    plt.tight_layout()
    plt.savefig(f'{feature}vsDependent_boxplot.png',bbox_inches='tight')
    plt.show();
```

```python
# checking for the missing values
data.isna().sum()
```

```python
# replace the missing values with the mean of the feature
for feature in data.columns:
    data[feature] = data[feature].fillna(data[feature].mean())
```

#### splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
x = data.drop('Default', axis=1)
y = data['Default']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state=1, stratify=y)
```

#### treating the dependent variable imbalance using SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy=0.6, random_state=1)
x_train_res, y_train_res = smote.fit_resample(x_train, y_train) 
```

```python
# check data balance after SMOTE
print(f'Before SMOTE: {y_train.value_counts()}')
print('-' * 60)
print(f'After SMOTE: {y_train_res.value_counts()}')
```

```python
# creating a dataset of the balanced data
data_SMOTE = pd.concat([x_train_res, y_train_res], axis=1)
```

```python
# splitting the balanced data into train and test set
x_res = data_SMOTE.drop('Default',axis=1)
y_res = data_SMOTE['Default']
x_train_res,x_test_res,y_train_res,y_test_res = train_test_split(x_res,y_res, test_size=0.3, random_state=1, stratify=y_res)
```

# Model Building 

### Pre-defind functions for model evaluation

```python
from sklearn.metrics import confusion_matrix, recall_score, accuracy_score, classification_report, roc_curve
```

```python
# predefined function for the confusion_matrix
def cm(actual, predicted):
    cm1 = confusion_matrix(actual, predicted)
    custom_cm1 = pd.DataFrame([[cm1[1,1],cm1[1,0]],[cm1[0,1],cm1[0,0]]],columns=['Predicted Positive','Predicted Negative'],index=['Actual Positive','Actual Negative'])
    return custom_cm1
```

```python
# pre-defined function for the ROC curve
def plot_roc_curve(actual,predicted):
    fpr,tpr,_ = roc_curve(actual,predicted)
    plt.plot(fpr,tpr,label='ROC Curve')
    plt.plot([0,1],[0,1],linestyle='--',color='grey')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.show();
```

```python
# pre-defined function for the variance inflation factor
def vif(data):
    vif_series = pd.Series([variance_inflation_factor(data.values,i) for i in range(data.shape[1])],
                           index=data.columns,
                           dtype=float)
    return vif_series
```
